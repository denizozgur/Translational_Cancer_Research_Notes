# Evidence of cancer is genetic disease

* Cancer is caused by agents that alter or mutate DNA, known as carcinogens. These mutations, which can range from small changes in a single DNA base to large chromosomal changes, accumulate over time. As we age, the chance of developing cancer increases because our DNA has more time to collect these mutations. This is why cancer is more common in older people and has become more prevalent as human lifespans have increased.

* Most mutations found in tumor cells are somatic, meaning they occur in body cells and are not inherited, but they are passed to daughter cells during cell division. This makes cancer a genetic disease at the cellular level. Only mutations in sperm or egg cells, called germ-line mutations, can be passed to offspring, but these rarely cause cancer directly.

* As cancer cells accumulate mutations, they continue to evolve, gaining a growth advantage over neighboring cells. This process mirrors Darwinian evolution, where random mutations lead to survival of the fittest cells, making cancer progression a natural and statistically inevitable outcome.


* Mutations accumulate in cells only when the cell’s defense mechanisms, such as DNA repair, fail. If DNA damage isn't repaired before the next cell division, it gets passed to daughter cells. Cells have several ways to repair DNA, and in severe cases, they trigger cell suicide to protect the body from transformation into cancerous cells. However, when these defense systems are overwhelmed, cells with harmful mutations can slip through.

### Three key processes control the number of cells in the body:

   * Cell proliferation, where cell division produces two new cells.
  * Programmed cell death (apoptosis), which eliminates damaged or unneeded cells.
  * Mutations that affect genes involved in these processes can upset the balance, leading to uncontrolled cell growth and potentially cancer.

# Cell growth , differentation regulation

* Normal genes that can become cancerous when mutated are called proto-oncogenes, which play essential roles in healthy cells. Every normal cell carries genes that have the potential to turn cancerous (oncogenic).

  * The cell cycle, the sequence of stages a cell goes through between divisions, consists of four stages:
    * G1
    * S phase (DNA synthesis)
    * G2
    * M phase (mitosis and cytokinesis).

* G1, S, and G2 are part of interphase, while most adult cells, which are not dividing, enter a resting state called G0.

Cells can re-enter the cycle when stimulated by growth factors, passing through a key checkpoint known as the G1 restriction point. Progress through the cell cycle is regulated by cyclins and cyclin-dependent kinases (cdks), which ensure proper timing and coordination.

# Carciogenesis promotion
 
* Tumor promotion can happen through various pathways, but a key factor is the formation of new blood vessels, known as angiogenesis. Studies show that tumors can only grow to about 1 mm in size without new capillaries. To expand further, cancer cells must produce angiogenic growth factors to stimulate blood vessel growth.

* The importance of new blood vessels is evident as cells near these vessels proliferate rapidly, while cell growth decreases the farther they are from the blood supply. This blood vessel development is crucial for supplying nutrients and oxygen to support the tumor's growth.

# Hormones Promotes

* In the study of rat mammary carcinogenesis, two key features highlight the role of hormones and cell development in cancer formation:

  *  Hormonal Influence: A single dose of the carcinogen DMBA can cause breast cancer, but only under specific conditions. If the ovarian hormone oestradiol is removed immediately after DMBA treatment, cancer development is blocked. However, reintroducing oestradiol reverses this effect, showing the hormone's critical role in promoting cancer.

  *  Timing and Development: DMBA is effective only when given at a particular stage, specifically when the mammary gland is developing. This suggests that cells are only vulnerable to cancer-causing agents at certain stages of development, linking carcinogenesis with the timing of cell growth and maturation.

# Loss genes may be carcinogenic 

* The terms 'repressor', 'antioncogene', 'recessive', and 'loss of function' describe genes that normally suppress cancer but, when inactivated, contribute to tumor formation.

* In experiments where normal and cancer cells are fused to create hybrids, the cells initially behave normally, but over time, they lose chromosomes containing these repressor genes and revert to cancer-like behavior. This highlights the role of these genes in preventing cancer.

* These experiments also show the inverse relationship between differentiation (specialization of cells) and cancer. For example, normal fibroblasts, which synthesize collagen (a marker of differentiation), retain this ability in hybrids but lose it when they become tumorigenic. This suggests that as cells lose their specialized functions, they are more prone to cancer development.

  - DNA directs protein production through RNA, which helps with cell structure and function.
  - **Unlike RNA and proteins** (short-lived), DNA must stay intact throughout life.

- ### **Mutation and DNA Damage**
  - DNA faces **environmental and internal attacks** that cause mutations or breaks.
  - Mutations in DNA can harm cells and their future copies (progeny).
  - **Cancer** involves changes in gene structure and expression.

- ### **Carcinogenesis and Mutations**
  - **Mutation accumulation** is a known cause of cancer development.
  - **Focus areas**:
    - Structure of genes.
    - Mutations occurring in cancer development.

- ### **Defense Mechanisms in Cells**
  - Cells detect and repair DNA damage, especially before cell division.
  - **Error detection during replication** prevents passing mutations to daughter cells.
  - **Cell cycle pause** may occur for DNA repair.
  - **Apoptosis** (cell death) is a last defense to stop spreading DNA damage and reduce cancer risk.

- ### **Focus on DNA Mutations and Repair**
  - **Mutation Causes**:
    - Exposure to carcinogens (cancer-causing agents).
  - **DNA Repair Systems**:
    - Systems exist to keep the genome intact and prevent tumor formation (tumorigenesis).


## DNA, Genes, and Gene Expression

- **DNA Structure and Genes**
  - Human DNA contains about **30,000 genes**.
  - DNA has a **double-helix** structure made of:
    - **Sugar**
    - **Phosphate**
    - **Nitrogenous bases**: adenine (A), guanine (G), cytosine (C), thymine (T)

- **Central Dogma of Molecular Biology**
  - Describes the flow of genetic information:
    - **DNA → RNA → Protein**
    - **DNA** is **transcribed** into RNA.
    - **RNA** is **translated** into protein.

- **Gene Expression**
  - Refers to the **transcription of a gene** into RNA.

- **Gene Structure**
  - Genes have two main functional parts:
    - **5'-end**: Contains the **promoter region**, which regulates gene expression.


## Gene Structure and Regulation

- **Promoter Region and Gene Expression**
  - **Promoter sequences** interact with proteins to control **RNA polymerase activity**.
  - These interactions determine **when and where** a gene is expressed.

- **Regulatory Regions**
  - Usually near the gene’s start, but can also be **downstream** or within **introns**.
  - **TATA Box** (sequence: TATAAAA):
    - Found near transcription start sites.
    - Binding of **TATA-binding protein (TBP)** is essential for starting transcription.

- **Response Elements (RE)**
  - Short DNA sequences in the promoter recognized by specific proteins.
  - **REs** help regulate gene expression by interacting with proteins.

- **Gene Structure**
  - Genes have **two main parts**:
    - **Regulatory region**: Controls when and where the gene is expressed.
    - **Coding region**: Contains the information to make a protein.


# Response Elements, Enhancer Elements, and Gene Transcription

- **Common Response Elements (RE)**
  - Genes with similar regulation often share common **response elements (REs)**.
  - **Example**: **Serum Response Element (SRE)**:
    - Sequence: **CCATATTAGG**
    - Found in genes that respond to serum.

- **Cell Cycle Regulation**
  - The **E2F response element** is essential for cell cycle control.
  - Found in promoters of **cyclin E** and **cyclin A** genes, which are key in cell cycle progression.

- **Enhancer Elements**
  - Additional **regulatory DNA sequences**.
  - **Position and orientation independent** relative to the promoter.
  - Important for **tissue-specific** and **stage-specific expression**.

- **Downstream Sequences**
  - Located after the **promoter**:
    - Transcribed into **RNA**.
    - **Exons** within this region are translated into **proteins**.


# Mutations

- ### **Carcinogens and Mutations**
  - Most **carcinogens** are also **mutagens**.
  - They cause mutations by:
    - **Modifying DNA** (e.g., forming DNA adducts).
    - **Causing chromosomal damage** (e.g., DNA strand breaks).

- ### **Types of Mutations**
  - **Transitions**: 
    - Substitute **one purine** for another purine (**adenine ↔ guanine**).
  - **Transversions**:
    - Substitute a **purine** for a **pyrimidine** (e.g., **adenine → cytosine**) or vice versa.
  - **Insertions and Deletions**:
    - Adding or removing nucleotides in DNA.
  - **Chromosomal Deletions**:
    - Loss of large DNA segments within a chromosome.

- **Causes of Base Substitutions**
  - **DNA Polymerase Errors**:
    - DNA polymerase is not always 100% accurate.
    - It may **insert the wrong nucleotide** during DNA synthesis.

  - **Base Modifications**:
    - **Oxidation** or **covalent additions** to bases can alter them.
    - Changes in **chromatin structure** may also lead to **misreading** of the DNA template by DNA polymerase.


- **Genetic Code**
  - **Triplet Code**: Read in sets of three bases, sequentially and non-overlapping.

- **Frameshift Mutations**
  - Caused by **insertion** or **deletion** of a base, which shifts the reading frame.
  - Often results in a **non-functional** or **truncated protein**.

- **Chromosomal Translocation**
  - Exchange of DNA segments between **different chromosomes**.
  - Leads to changes in the **DNA base sequence**.

- **Relevance to Carcinogenesis**
  - Mutations like frameshifts and translocations affect genes involved in:
    - **Growth regulation**
    - **Differentiation**
    - **Apoptosis** (programmed cell death)
  - These mutations play a role in **cancer development**.

- **Mutation Location and Effect**
  - Mutations can theoretically occur **anywhere within a gene**.
  - The **location** of a mutation affects whether it will:
    - **Provide a growth advantage** to the cell.
    - Contribute to **carcinogenesis** (cancer development).

- **Example: Cyclin Protein Mutation**
  - Mutation in a **cyclin protein** may change its **shape (conformation)**.
  - This could lead to **unregulated cell cycle progression** (a cancer risk).
  - In contrast, other mutations might **not affect** protein structure or function, having no impact on growth.

- **Mutation Consequences Based on Gene Region**
  - The **impact of a mutation** depends on its location within a gene’s two main parts:
    - **Promoter Region**
      - Mutations here can alter **gene regulation**.
      - May affect **levels** or **timing/location** of gene expression.
      - Possible outcomes:
        - **Over- or under-expression** of the protein.
        - Protein appears at the **wrong time** or **wrong place**.
    - **Coding Region**
      - Mutations here can change **protein structure**, affecting function.
      - May cause **truncation** (e.g., by introducing a stop codon), leading to:
        - Loss of protein function entirely.

# Carcinogens and Radiation

- ### **Carcinogens and Cancer Biology**
  - **Carcinogens** cause cancer by inducing mutations.
  - Identifying these **mutagens** and the specific mutations they cause has been key to understanding **carcinogenesis**.
  - Carcinogens include:
    - **Radiation**
    - **Chemicals**
    - **Infectious pathogens**
    - **Endogenous reactions**

- **Radiation as a Carcinogen**
  - Radiation is energy that can travel in **waves** or as **atomic particles**.

- ### **Types of Radiation**
  - **Energy Waves**:
    - **Gamma (γ) rays**: High-energy, similar to **X-rays**.
  - **Atomic Particles**:
    - **Alpha (α) particles**: Composed of **2 protons** and **2 neutrons**.
    - **Beta (β) particles**: Composed of **electrons**.

- ### **Electromagnetic Radiation**
  - **Electromagnetic Radiation**: Moves as waves with **peaks** and **troughs**.
    - The distance between peaks/troughs is called **wavelength**.
  - **High-energy radiation** (e.g., **cosmic radiation**) has a **short wavelength**.
  - **Low-energy radiation** (e.g., **radio waves**) has a **long wavelength**.

- ### **The Electromagnetic Spectrum**
  - The spectrum includes various types of radiation, from long to short wavelengths:
    - **Cosmic radiation**: Short wavelength, high-energy.
    - **X-rays & Gamma rays**: Short wavelengths with high energy.
    - **Ultraviolet (UV) Radiation**: Higher energy (shorter wavelength) than visible light.
    - **Visible Light**: Wavelengths detectable by the human eye.

## Radiation and DNA Damage
- **Radiation as Carcinogens**
  - **Radiation types** (both energy waves and atomic particles) can **damage DNA** and act as carcinogens.

- **Energy and DNA Damage**
  - The **amount of energy** released by a radiation source affects the **extent of DNA damage**.
  - **High-energy radiation** causes **more damage** compared to radiation with a **lower energy release rate**.

- **Classes of Radiation That Cause DNA Damage**
  - **Ionizing Radiation**: Can directly damage DNA by removing electrons.
  - **UV Radiation**: Can cause **thymine dimers** and other DNA distortions.


# Ionizing Radiation and DNA Damage

- **Ionizing Radiation Types**
  - **Alpha and Beta Particles** (atomic particles)
  - **Gamma Rays** (energy waves)

- **How Ionizing Radiation Works**
  - High-energy radiation (e.g., γ-rays) strikes molecules and displaces **electrons** from atoms.
  - Loss of electrons **creates ions**, which are molecules with an **electrical charge**.
  - **Ionizing radiation** causes molecules to become **ions** and can damage DNA.

- **DNA Damage from Ionizing Radiation**
  - **Direct Damage**: Ionization of atoms in DNA.
  - **Indirect Damage**: Interaction with **water molecules** (radiolysis), which generates **reactive oxygen species (ROS)**.

- **Reactive Oxygen Species (ROS)**
  - **ROS Formation**: Sequence of reactions leading to three ROS:
    - **Hydroxyl Radical (-OH)**: Highly reactive, triggers chain reactions.
    - **Hydrogen Peroxide (H₂O₂)**: Less reactive, but more dangerous to DNA.
    - **Superoxide Radical (O₂-)**: Less reactive.
  - **Hydrogen Peroxide** can travel into the **nucleus** and damage DNA.

- **DNA Damage from ROS**
  - **Oxidation** of DNA by ROS (e.g., removal of electrons) causes mutations.
  - **8-oxoguanine** is a common oxidized base. It leads to a **G → T** mutation during DNA replication.

- **Exposure to Ionizing Radiation**
  - **Exposure Levels**:
    - **High-altitude flights**: 0.005-0.01 mSv per hour.
    - **Chest X-ray**: 0.1 mSv.
  - **Cancer Risk**: Accumulation of radiation exposure over time is linked to cancer risk.

- **Key Findings from Studies**
  - **Double-strand DNA breaks** are a significant result of ionizing radiation-induced cancer.
  - **Atomic Bomb Survivors**:
    1. **Leukemia** is the most common cancer from radiation.
    2. **Age** is a factor: Children exposed to radiation have a higher risk.
    3. **Solid cancers** increase linearly with radiation dose. The risk is **47% higher per sievert** for those exposed at 30 years old when they reach 70 years old.


# Ultraviolet (UV) Radiation and DNA Damage

- **Types of UV Radiation**
  - **UV-A (320-380 nm)**: Least harmful.
  - **UV-B (290-320 nm)**: Most effective carcinogen. (because of our atmosphere)
  - **UV-C (200-290 nm)**: Highly dangerous but mostly absorbed by the atmosphere.

- **How UV Radiation Affects DNA**
  - **Conjugated double bonds** in nitrogenous bases of DNA absorb UV radiation.
  - **UVB radiation** causes **pyrimidine photoproducts**:
    - **Cyclobutane Pyrimidine Dimers (CPDs)**: Most common UV-induced DNA damage, formed 20-40 times more than other photoproducts.
    - **Pyrimidone photoproducts**: Mimic an **abasic site**, which is easier to repair than CPDs.

- **Effects of Pyrimidine Dimers**
  - Pyrimidine dimers cause a **bend** in the DNA helix.
  - **DNA polymerase** cannot read the DNA template at the dimer site, leading to errors.
  - DNA polymerase **incorporates "A" residues** preferentially, causing mutations:
    - **TT dimers** are usually restored.
    - **TC and CC dimers** result in **transitions** (e.g., **TC → TT**, **CC → TT**).

- **UVB and Skin Cancer**
  - **Cyclobutane Pyrimidine Dimers (CPDs)** are responsible for **80% of UVB-induced mutations**.
  - UV-induced mutations, especially **pyrimidine dimers**, are a signature of **skin cancer** and are rarely found in other cancers.

- **UVA Radiation and DNA Damage**
  - **UVA** causes **indirect DNA damage** through **free radicals**.
  - **Water molecules** are fragmented by UVA, generating **reactive oxygen species (ROS)** that damage DNA.
  - **G → T transversions** are typical of UVA-induced damage.


# Chemical Carcinogens

- **Mechanism of Action**
  - Carcinogens often act by forming **electrophilic** (electron-deficient) compounds that react with **nucleophilic** sites (electron-donating) in purine and pyrimidine rings of DNA.
  - Some carcinogens act **directly on DNA**, while others become active **only after metabolism**, forming **ultimate carcinogens** that cause DNA damage.

- **Cytochrome P450 Enzymes**
  - **Cytochrome P450 enzymes** in the liver are crucial for the metabolism of chemicals, activating carcinogens to their **ultimate form**.
  - Genetic differences and enzyme expression (e.g., **CYP1A1** enzyme) contribute to **individual variations** in responses to carcinogens, such as those seen in smokers.

- **Types of Chemical Carcinogens**
  - Carcinogens are grouped into 10 classes:
    1. **Polycyclic Aromatic Hydrocarbons (PAHs)**
    2. **Aromatic Amines**
    3. **Azo Dyes**
    4. **Nitrosamines and Nitrosamides**
    5. **Hydrazo and Azoxy Compounds**
    6. **Carbamates**
    7. **Halogenated Compounds**
    8. **Natural Products**
    9. **Inorganic Carcinogens**
    10. **Miscellaneous Compounds** (alkylating agents, aldehydes, phenolics)
  - Focusing on **4 Major Classes**:
    - **Polycyclic Aromatic Hydrocarbons (PAHs)**
    - **Aromatic Amines**
    - **Nitrosamines**
    - **Alkylating Agents**

- **DNA Damage by Chemical Carcinogens**
  - Carcinogens covalently add **functional groups** to DNA, forming **DNA adducts**.
  - These adducts **distort the DNA helix**, leading to errors during DNA replication.
  - The errors result in **mutations**, which can initiate **carcinogenesis**.


## 1. OGG1 Gene Mutations:
- Mutations in the **OGG1** gene (responsible for repairing 8-oxoguanine) have not been found in tumors.
- Mutations in the **MYH** gene may cause **multiple colorectal adenoma syndrome**.

## 2. Base Excision Repair (BER):
- DNA **glycosylase** enzymes cleave the altered base and remove base-free sugars.
- **Short patch repair**: A single nucleotide gap is filled by DNA polymerase β and sealed by ligase.
- **Long patch repair**: Strand-displacing polymerases extend the gap, and flap endonuclease removes displaced strands.

## 3. Mismatch Repair:
- Corrects replication errors, including insertions, deletions, and mismatches.
- Involves recognition by **HMSH2/6**, **hMSH2/3**, and recruitment of **hMLH1/hPMS2**, followed by endonucleases and exonucleases.

## 4. Hereditary Non-Polyposis Colorectal Cancer (HNPCC):
- Associated with germline mutations in **hMLH1** or **hMSH2**, leading to complete mismatch repair loss.
- Increases vulnerability to mutations.

## 5. Recombinational Repair:
- **Homologous recombination** and **non-homologous end joining** repair double-strand breaks.
- Involves **ATM kinase**, **RAD50/MRE11/NBS1 complex**, **BRCA2**, **RAD51**, and **RAD52**.
- **RAD51** aids in homologous recombination by exchanging sequences from a sister chromatid.


## 1. Alkylating Agents and Platinum-Based Drugs:
- **Cisplatin** replaces CI with water in plasma and cytosol before binding to the N7 position of guanine and adenine in DNA.
- **GG**, **AG**, and **GXG** (where X is any base) adducts make up over 90% of total adducts.
- The DNA damage caused triggers **apoptosis**.
- **Cisplatin** has major impacts on cancers like ovarian cancer but is associated with irreversible kidney damage.
- **Carboplatin** is a less toxic platinum analog.

## 2. Antimetabolites:
- Antimetabolites mimic endogenous molecules (e.g., nitrogenous bases of DNA) and inhibit nucleic acid synthesis.

### 2.1. **5-Fluorouracil (5-FU)**:
- A derivative of **uracil**, converted into **fluorodeoxyuridylate (F-dUMP)**.
- **F-dUMP** competes with **dUMP** for thymidylate synthase, acting as a **suicide inhibitor**.
- This results in depletion of **dTMP** and **dTTP** pools, causing severe compromise in DNA synthesis in rapidly dividing cells.

### 2.2. **Methotrexate**:
- **Methotrexate** is an analog of **dihydrofolate** and inhibits **dihydrofolate reductase**, preventing regeneration of tetrahydrofolate needed for thymidylate synthase.

## 3. Organic Drugs:
### 3.1. **Doxorubicin**:
- An anthracycline antibiotic that inhibits **topoisomerase II**, causing torsional stress during DNA replication.
- Can cause **cardiac damage**, but compounds like **ICRF-187** are being studied to prevent toxicity.
- Used for solid tumors (e.g., breast, lung cancer).

### 3.2. **Vincristine and Vinblastine**:
- Derived from the Madagascar periwinkle plant, they bind to **tubulin** and prevent microtubule assembly.
- **Paclitaxel** stabilizes microtubules, preventing their depolymerization, opposite to vincristine/vinblastine's action.

## 4. Radiation Therapy:
- **Radiation therapy** is received by approximately **60% of cancer patients** in the USA.
- **Ionizing radiation** reacts with water inside cells to generate **ROS**, which damage DNA.
- **Apoptosis** is induced in cells with significant DNA damage.
  
### 4.1. Oxygen’s Role:
- Oxygen enhances radiation-induced damage by aiding in the formation of **double-strand breaks**.
- **Hypoxia** zones in tumors can influence the effectiveness of radiation therapy.

### 4.2. Modern Techniques:
- **MRI** and **CT** scans are used to produce **3D images** of the tumor, improving the precision of tumor targeting during radiation therapy.


## 1. Drug Resistance:
- Cells can become resistant to drugs through various mechanisms:
  - **Increasing the efflux of the drug**
  - **Decreasing the intake of the drug**
  - **Increasing the number of target molecules** within the cell
  - **Alterations in drug metabolism**
  - **Enhanced DNA repair processes**

## 2. Drug Efflux and Resistance:
- **Efflux of drugs** is regulated at the cell surface.
- **ATP-dependent transporters** are involved in the movement of nutrients and other molecules across membranes.
- **Multi-drug resistance gene (MDR1)** encodes **P-glycoprotein (P-gp)**, a member of this transporter family, which can pump out various chemotherapeutic drugs such as **doxorubicin**, **vinblastine**, and **taxol**.
- P-gp works by hydrolyzing **ATP** to change conformation, expelling the drug from the cell. The transporter is recycled with a second ATP hydrolysis.

## 3. Drug Influx and Resistance:
- Some drugs rely on **specific transporters** to enter cells. Mutations in these transporters may reduce drug uptake.
- **Resistance to methotrexate** can occur due to mutations in the **folate transporter**.
- **Gene amplification** can increase the number of drug target molecules. For example, amplification of the **DHFR** gene increases resistance to methotrexate.
  
## 4. DNA Repair and Resistance:
- Increased **DNA repair** mechanisms, such as elevated **alkyltransferase activity**, can lead to resistance against **alkylating agents** like **doxorubicin**.
- Alkyltransferase levels are highly variable in different tumors.
